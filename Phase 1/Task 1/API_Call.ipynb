```python
import pandas as pd
import requests
import json
```

```python
api_url = "https://api.binance.com/api/v3/ticker/24hr"
```

```python
response = requests.get(api_url)
data = json.loads(response.text)
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.to_csv("binance_24hr_ticker_data.csv", index=False)
```
